```python
exchange_rates_df = pd.read_csv("exchange_rates.csv", index_col=0)
```

```python
exchange_rates_df.head()
```

```python
exchange_rate = exchange_rates_df.loc['GBP']
exchange_rate
```

```python
import json
```

```python
with open('bank_market_cap_1.json', 'r') as f:
  data1 = json.load(f)
df1 = pd.DataFrame(data1)
df1.head()
```

```python
with open('bank_market_cap_2.json', 'r') as f:
  data2 = json.load(f)
df2 = pd.DataFrame(data2)
df2.head()
```

```python
df = df1.append(df2, ignore_index=True)
df.head()
```

## Transform (market cap from USD to GBP).
Round it to 3 decimal places. Also change the column name to GBP$ Billion

```python
df.rename(columns={list(df)[1]: "Market Cap (GBP$ Billion)"}, inplace=True)
```

```python
df.head()
```

```python
for i, value in enumerate(df['Market Cap (GBP$ Billion)']):
  usd_to_gdp = round(float(value) / float(exchange_rate[0]), 3)
  df['Market Cap (GBP$ Billion)'][i] = usd_to_gdp
```

```python
df.head()
```

```python
#load into a csv file
df.to_csv("bank_market_cap_gbp.csv", index=False)
```
